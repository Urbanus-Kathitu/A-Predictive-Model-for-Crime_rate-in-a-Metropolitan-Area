## Import Libraries

```python
# Let's first load the required libraries.

# Load libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder # For encoding purposes
from sklearn.tree import  DecisionTreeRegressor # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
# Let's first load the required Standard Metropolitan Areas dataset using pandas read CSV function.
Areas_Data = pd.read_csv("Standard Metropolitan Areas Data - train_data - data.csv")
```

#FEATURE SELECTION
Here, you need to divide given columns into two types of variables dependent(or target variable) and independent variable(or feature variables).

```python
Y = Areas_Data.crime_rate # Target variable
```

```python
#split dataset in features and target variable
Area_features = ['region', 'percent_city', 'work_force','graduates','income']
```

```python
X = Areas_Data[Area_features]
```

```python
X.describe()
```

```python
# Display the head of the dataset
X.head()
```

#Splitting Data

To understand model performance, dividing the dataset into a training set and a test set is a good strategy.

Let's split the dataset by using function train_test_split(). You need to pass 3 parameters features, target, and test_set size.

```python
# Split dataset into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=59) # 70% training and 30% test
```


##Building Your Model

Now we are finally ready, and we can train the model.

First, we need to import our model - A decision tree classifier (again, using the sklearn library).

Then we would feed the model both with the data (X_train) and the answers for that data (y_train)

```python
# Create Decision Tree classifer object
clf = DecisionTreeRegressor()

# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
Crime_model = clf.predict(X_test)
```

```python
# Define model. Specify a number for random_state to ensure same results each run
Crime_model = DecisionTreeRegressor(random_state=1)

# Fit model
Crime_model.fit(X, Y)
```

```python
# Display the results of the colums that infulence the crime rate the most in %
print(X.head())

print(Crime_model.predict(X.head()))
```

##Visualizing Decision Trees

```python
# You can use Scikit-learn's export_graphviz function for display the tree within a Jupyter notebook. 
from sklearn.tree import export_graphviz
import graphviz
```

```python
dot_data= export_graphviz(Crime_model, out_file=None, 
                    feature_names=X_test.columns,  
                      class_names=['0', '1'],  
                      filled=True, rounded=True,  
                      special_characters=True) 
```

```python
graph1=graphviz.Source(dot_data)
```
